```python
import pandas as pd
import numpy as np


np.random.seed(42)

data = {
    "Student_ID": range(1, 101),
    "Study_Hours": np.random.normal(5, 2, 100),
    "Attendance": np.random.normal(75, 15, 100),
    "Marks": np.random.normal(65, 20, 100),
    "Gender": np.random.choice(["Male", "Female"], 100),
    "Department": np.random.choice(["CS", "IT", "EXTC"], 100)
}

df = pd.DataFrame(data)


df.loc[5, 'Marks'] = np.nan
df.loc[10, 'Attendance'] = np.nan
df.loc[20, 'Gender'] = None

df.head()
```

```python
df.isnull().sum()
```

```python
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
df['Attendance'] = df['Attendance'].fillna(df['Attendance'].mean())
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df['Study_Hours'] = df['Study_Hours'].clip(lower=0)
df['Attendance'] = df['Attendance'].clip(0, 100)
df['Marks'] = df['Marks'].clip(0, 100)
```

```python
Q1 = df['Marks'].quantile(0.25)
Q3 = df['Marks'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Marks'] < lower) | (df['Marks'] > upper)]
print(outliers)
```

```python
df['Marks'] = np.where(df['Marks'] > upper, upper, df['Marks'])
df['Marks'] = np.where(df['Marks'] < lower, lower, df['Marks'])
```

```python
df['Marks'].skew()
```

```python
df['Marks_log'] = np.log1p(df['Marks'])
```

```python
df['Marks_scaled'] = (df['Marks'] - df['Marks'].mean()) / df['Marks'].std()
```
