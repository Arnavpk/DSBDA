```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("iris.csv")

df.columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species']

# Show first rows
print("First 5 rows:\n", df.head())
```

```python
# 📊 Data types
print("\nData Types:\n", df.dtypes)
```

```python
df.hist(figsize=(10,8))
plt.suptitle("Histogram of Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df.iloc[:, :-1])
plt.title("Boxplot of Features")
plt.show()
```

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species']

for feature in features:
    plt.figure()
    sns.boxplot(x='Species', y=feature, data=df)
    plt.title(f"{feature} vs Species")
    plt.show()
```

```python
print("\nInference:")
print("- All features are numeric except species.")
print("- Petal features clearly separate species.")
print("- Sepal width shows some outliers.")
```
